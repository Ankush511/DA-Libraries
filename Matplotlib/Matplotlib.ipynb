# 1) Importing Libraries

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## a) Generating a Sine and Cosine wave

```python
x = np.linspace(0, 4*np.pi, 100)

# defining wave
sin_y = np.sin(x)
cos_y = np.cos(x)

# plotting the wave
plt.plot(x, sin_y, label='Sine')
plt.plot(x, cos_y, label='Cosine')

plt.legend()
plt.title('Sine & Cosine wave')
plt.show()
```

```python
x = np.linspace(0, 4*np.pi, 100)

# defining wave
sin_y = np.sin(x)
cos_y = np.cos(x)

# plotting the wave
plt.plot(x, sin_y, '*')
plt.plot(x, cos_y, '--')

plt.title('Sine & Cosine wave')

plt.show()
```

# 4) Different types of plots

## a) Line Plot

```python
# Single line plot
x = np.linspace(0,4*np.pi,100)
y = np.sin(x)

plt.style.use('Solarize_Light2')

plt.plot(x, y)

plt.title('Sine wave')

plt.show()
```

```python
# Multiple line plot
x = np.linspace(0,4*np.pi,100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(15,5))

plt.style.use('Solarize_Light2')

plt.plot(x, y_sin, color='red', lw=1, label='sin')
plt.plot(x, y_cos, color='blue', lw=1, label='cos')

plt.title('Sine & Cosine wave')
plt.legend()

plt.show()
```

### Using different linestyles

```python
x = np.linspace(0,20,21, dtype='int')

# using different linestyle
plt.figure(figsize=(15,8))

plt.plot(x, x+0, ls='solid', label='solid')
plt.plot(x, x+1, ls='dashed', label='dashed')
plt.plot(x, x+2, ls='dashdot', label='dashdot')
plt.plot(x, x+3, ls='dotted', label='dotted')

plt.legend(loc='lower right')
plt.show()
```

### Adjusting the plot axes

```python
# using plt.xlim and plt.ylim
x = np.linspace(0, 4*np.pi, 100)

plt.xlim(0,15)
plt.ylim(-2.5,2.5)

sin_y = np.sin(x)
plt.plot(x, sin_y)
plt.title('Sine wave')

plt.show()
```

```python
# we can also use plt.axis([xlim,ylim])
x = np.linspace(0, 4*np.pi, 100)

plt.axis([0,15,-2,2])

sin_y = np.sin(x)
plt.plot(x, sin_y)
plt.title('Sine wave')

plt.show()
```

- using plt.axis('tight') or plt.axis('equal') we can avoid manual graph adjusting

```python
# using .set to combine all of these commands
x = np.linspace(0,4*np.pi,100)
fig = plt.axes()

fig.plot(x,np.sin(x))
fig.set(xlim=(0,10),ylim=(-2,2),xlabel='x',ylabel='sin(y)',title='Sine Wave')

fig.plot()
```

## b) Scatter Plot

### Single scatter plot

```python
n = 500
x = np.random.randn(n)
y = np.random.randn(n)

plt.scatter(x,y, c='red', alpha=.5, s=30)

plt.title('Scatter Plot')
plt.show()
```

### Multi Scatter Plot

```python
n = 1000

x1 = np.random.randn(n)
y1 = np.random.randn(n)

x2 = np.random.randn(n) + 2
y2 = np.random.randn(n) + 2

plt.scatter(x1,y1,c='red',alpha=0.7,s=20,label='Group-1')
plt.scatter(x2,y2,c='blue',alpha=0.7,s=20,label='Group-2')

plt.legend()
plt.title('Multi Scatter Plot')
plt.show()
```

### On real world datasets

```python
from sklearn.datasets import load_iris

iris = load_iris()
x = iris.data
y = iris.target

plt.scatter(x[:,0],x[:,1],c='blue',alpha=0.8,s=50)

plt.xlabel(iris['feature_names'][0])
plt.ylabel(iris['feature_names'][1])

plt.show()
```

```python
# using markers
plt.scatter(x[:,2],x[:,3],c='blue',alpha=0.8,s=50,marker = 'x')

plt.xlabel(iris['feature_names'][0])
plt.ylabel(iris['feature_names'][1])
```

#### using more properties on scatter plot

```python
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = rng.rand(100) * 1000

plt.scatter(x,y, c=colors, s=sizes, alpha=.7, cmap='viridis', edgecolors='black')

plt.title('Random Scatter Plot')
plt.xlabel('x values')
plt.ylabel('y values')

plt.colorbar()

plt.show()
```

#### Using above properties on Iris dataset

```python
iris = load_iris()
features = iris.data

plt.scatter(features[:,0],features[:,1], alpha=0.7, s=features[:,3]*100, c=iris.target, cmap='viridis', edgecolors='black')

plt.colorbar()

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

plt.show()
```

## c) Histogram and Binnings

```python
# normal histogram
data = np.random.randn(100)
plt.hist(data)

plt.show()
```

### histogram with defined bins,edgecolor,color,alpha,histtype

```python
data = np.random.randn(1000)
plt.hist(data, bins=25, edgecolor='black', color='green', alpha=.7, histtype='stepfilled')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Random Histogram')

plt.show()
```

## d) Pie Charts

```python
# Normal Pie chart 
cars = ['Audi','Tesla','Ford','Mercedes','Jaguar','BMW']
data = [210,25,100,245,170,195]

plt.figure(figsize=(8,8))

plt.pie(data, labels=cars)
plt.show()
```

### Basic operations on Pie Chart

```python
cars = ['Audi','Tesla','Ford','Mercedes','Jaguar','BMW']
data = [210,75,100,245,170,195]
explode = [0,0,.1,0,0,0]

plt.figure(figsize=(8,8))

plt.pie(data, labels=cars, explode=explode, shadow=False, autopct='%0.2f%%')
plt.show()
```

### Adding wedge props

```python
plt.figure(figsize=(8,8))

plt.pie(data, labels=cars, explode=explode, shadow=False, autopct='%0.2f%%', wedgeprops={
    'edgecolor' : 'black', 'antialiased' : True, 'lw' : 2
})
plt.show()
```

## e) Multiple Subplots

```python
# Multiple plots using index

x = np.linspace(0,10,50)

fig, ax = plt.subplots(2,2)

y = np.sin(x)
ax[0][0].plot(x,y)
ax[0][0].set_title('Sin wave')

y = np.cos(x)
ax[0][1].plot(x,y)
ax[0][1].set_title('Cos wave')

y = np.tan(x)
ax[1][0].plot(x,y)
ax[1][0].set_title('Tan wave')

y = np.random.randn(1000)
ax[1][1].hist(y, bins=30)
ax[1][1].set_title('Random histogarm')


plt.tight_layout()
plt.show()
```

```python
# Multiple plots with loops

plt.subplots_adjust(hspace=0.4,wspace=0.4)

for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5,0.5, str((2,3,i)) ,ha='center', fontsize = 18)
```

```python
# Multiple subplots another method

x = np.linspace(0,10,100)
y = np.sin(x)

fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(18,6))

ax1.plot(x,y)
ax2.plot(x,y**2)
ax3.plot(x,y**3)

ax1.set_title('Sin Wave')
ax2.set_title('Sin Wave - Square')
ax3.set_title('Sin Wave - Cube')

plt.tight_layout()
plt.show()
```
